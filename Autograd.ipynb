```python
import torch
```

```python
x = torch.tensor(3.0,requires_grad=True) # pytorch understand tha we r trying to. calculate derivative
```

```python
y = x**2 ## forward pass
x
```

```python
y
```

```python
y.backward() ## gradient calculate vayo
```
